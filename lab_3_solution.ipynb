```python
file1['vehicle_size'] = file1['vehicle_size'].fillna('Medsize')
```

6. Datetime format - Extract the months from the dataset and store in a separate column. 
Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python
file1['effective_to_date'] = pd.to_datetime(file1['effective_to_date'], errors='coerce')
file1
```

```python
file1['month'] = file1['effective_to_date'].dt.month  # year
# file1['month'] = pd.DatetimeIndex(file1['effective_to_date']).month   # year
```

```python
file1 = file1[(file1['month'] >= 1) & (file1['month'] <= 3)]     
# in dataframe use '&' '|' instead of 'and' 'or'
```
